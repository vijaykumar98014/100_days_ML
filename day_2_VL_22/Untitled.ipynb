```python
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
titanic = pd.read_csv('train.csv')
titanic.head()
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
#Scattered num-num

sns.scatterplot(x='total_bill', y='tip', data=tips,color = 'red')
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips,color = 'red',hue = 'sex')
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips,color = 'red',hue = 'sex',style = 'smoker')
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips,color = 'red',hue = 'sex',style = 'smoker',size = 'size')
```

```python
titanic.head()
```

```python
sns.barplot(x='Pclass', y='Fare', data=titanic,hue = 'Sex')
```

```python
#Box plot 

sns.boxplot(x = 'Sex',y = 'Age',data = titanic,hue = 'Survived')
```

```python
# Dis plot

sns.distplot( titanic[titanic['Survived'] == 0]['Age'])
```

```python
#cat-catigorical 

sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
(titanic.groupby('Pclass')['Survived'].mean()* 100)
```

```python
#cluster map

sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```
